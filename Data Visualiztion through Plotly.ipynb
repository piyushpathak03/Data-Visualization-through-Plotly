```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
import cufflinks as cf
```

## iplot() can visualize all types of 2d and 3d graph
* line
* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histogram
* bubble
* scatter3d
* bubble 3d

```python
py.offline.init_notebook_mode(connected=True)  #connect offline
cf.go_offline()
```

```python
#Create Dataframe
df = pd.DataFrame(np.random.randn(120,3), columns = ['A', 'B', 'C'])
df.head()
df['A'] = df['A'].cumsum() + 25
df['B'] = df['B'].cumsum() + 25
df['C'] = df['C'].cumsum() + 25
```

```python
df.head()
```

```python
#Ploting through iplot
df.iplot()
```

```python
#compare with matlplotlib
df.plot()     
plt.show()   ## through matplotlib
```

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 20,colors='hotpink',theme='henanigans')
```

```python
df.iplot(x = 'B', y = 'C', mode = 'markers', size = 20,colors='red',theme='henanigans')
```

```python
cf.getThemes()
```

```python
cf.set_config_file(theme='ggplot')
df.iplot(kind = 'bar', barmode='stack', bargap=0.5,theme='solar')
```

```python
df.iplot(kind = 'bar', barmode='stack', bargap=0.8)
```

```python
df.iplot(kind = 'barh', barmode='stack', bargap=0.5)
```

```python
df.iplot(kind = 'box',theme='solar')
```

```python
df.iplot(kind = 'area',theme='solar')
```

```python
df[['A', 'B']].iplot(kind = 'spread')
```

```python
df.iplot(kind='hist', bins = 25, barmode = 'overlay', bargap=0.5)
```

# 3d Plot

```python

df3 = pd.DataFrame({'X': [10,20,30,20,10], 'Y': [10, 20, 30, 20, 10], 'Z': [10, 20, 30, 20, 10]})
df3.head()
```

```python
df3.iplot(kind='surface', colorscale='rdylbu')   #Always use kind='surface for 3d graph'
```

```python
#To generate new data records we use cufflinks.datagen : a datagenerator
cf.datagen.sinwave(20, 0.33).head()
```

```python
#Plot 3D graph
cf.datagen.sinwave(20, 0.33).iplot(kind = 'surface')
```

```python
# generata 3d random data to plot scatter chart
cf.datagen.scatter3d(2, 120, mode = 'stocks').head()
```

```python
cf.datagen.scatter3d(2, 120, mode = 'stocks').iplot(kind = 'scatter3d', x = 'x', y= 'y', z = 'z')
```

```python
#3d bubble chart
cf.datagen.bubble3d(5,4,mode='stocks').iplot(kind='bubble3d',x='x',y='y',z='z', size='size')
```

```python
#Heatmap
cf.datagen.heatmap(20,20).iplot(kind = 'heatmap', colorscale='spectral' ,title='Cufflinks - Heatmap')
```
